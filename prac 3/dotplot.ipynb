```python
# Matrix Addition with OpenMP

import matplotlib.pyplot as plt

# Data for the plot
threads = [2, 4, 8, 12]
execution_time = [0.010084, 0.004367, 0.004567, 0.003296]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(threads, execution_time, 'bo-', markersize=8, linewidth=2)

# Add labels and title
plt.xlabel('Number of Threads', fontsize=14)
plt.ylabel('Execution Time (seconds)', fontsize=14)
plt.title('Execution Time vs. Number of Threads (Matrix Size: 2000x2000)', fontsize=16)

# Annotate the points with execution times
for i in range(len(threads)):
    plt.text(threads[i], execution_time[i], f'{execution_time[i]:.6f}', fontsize=12, ha='right')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
